```python
import pandas as pd

# display all
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)

path = "/Users/matteovitali/Desktop/polymarket-trading-bot/experiments/interesting/paper/BrownianStrategy_20260418_101103/trades.csv"
```

```python
trades = pd.read_csv(path)
actual = trades[trades["exit_reason"] != "no_trade"].copy()

actual.sort_values("entry_time")
```

Check summary is correct

```python
wins = actual[actual["pnl"] > 0]
losses = actual[actual["pnl"] <= 0]

print(f"Total trades:  {len(actual)}")
print(f"Win rate:      {len(wins)/len(actual):.2%}")
print(f"Total PnL:     {actual['pnl'].sum():.4f}")
print(f"Avg PnL:       {actual['pnl'].mean():.4f}   !!!!!!")
print(f"Avg win PnL:   {wins['pnl'].mean():.4f}")
print(f"Avg loss PnL:  {losses['pnl'].mean():.4f}") # should always be -1.00 if no stop-loss
```

#### Analyze perfromance

```python
# calculate pnl for trades with entry time in a particular span
span = [21, 300]

filtered_trades = actual[(actual["entry_time"] >= span[0]) & (actual["entry_time"] <= span[1])]

print(f"Total trades:  {len(filtered_trades)}")
print(f"Win rate:      {len(filtered_trades[filtered_trades['pnl'] > 0]) / len(filtered_trades):.2%}")
print(f"Total PnL:     {filtered_trades['pnl'].sum():.4f}")
print(f"Avg PnL:       {filtered_trades['pnl'].mean():.4f}  !!!!!!")
print(f"Avg win PnL:   {filtered_trades[filtered_trades['pnl'] > 0]['pnl'].mean():.4f}")
print(f"Avg loss PnL:  {filtered_trades[filtered_trades['pnl'] <= 0]['pnl'].mean():.4f}")
```

```python
# do a grid search that search maximum avg pnl for different spans that are at least 60 seconds long
best_span = None
best_avg_pnl = float("-inf")
for start in range(0, 300, 5):
    for end in range(start + 60, 300, 5):
        span = [start, end]
        filtered_trades = actual[(actual["entry_time"] >= span[0]) & (actual["entry_time"] <= span[1])]
        avg_pnl = filtered_trades['pnl'].mean()
        print(f"Span: {span}, Avg PnL: {avg_pnl:.4f}")
        if avg_pnl > best_avg_pnl:
            best_avg_pnl = avg_pnl
            best_span = span

print(f"Best span: {best_span}, Best avg PnL: {best_avg_pnl:.4f}")
```
